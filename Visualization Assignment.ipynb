# Crime in San Francisco

San Francisco is a city located on western coast of the United States. It is home to around 850,000 people. City of San Francisco is also well known as the hub of information technology and accommodates many billionaires. Income gap between various groups is very high in the city. Such income gaps can lead to crime provocative behaviour. Recently, the police department of the city made criminal data public. This post contains the visual representation of the criminal records to better understand the crime in the city of San Francisco.

All the visualization will be created using Python. In order to create the visualization, we will need to import the necessary packages. Python code given below imports the required packages for the visualization.

```python
## To execute the plots in the IPython notebook.
%matplotlib inline

## Import the required packages for the visualization.
import csv
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.cm as cm
import numpy as np
import scipy as sp

## Read Data from CSV file and store it into the dataframe.
df = pd.read_csv('sfd.csv')
```

## Overview of Crime in the City

The visualization below illustrates the most common categories of the crime that takes place in the city. Most common type of crime that occurs around the city is act of robbery. More than half of the offence in the city are theft related. Other commonly occurring categories of the crime include non-criminal, assault and vehicle theft. The visualization can also help to determine the rarest type of criminal activities such as bribery, gambling and pornography.


The code given below prints the crime related data in tabular format. Table contains 2 columns namely category of the crime and t frequency associated with each type of the category of the crime. After the table, the visualization is created. Visualization is a bar graph and height of the bar denotes the frequency of the crime.

```python
## Create a new Data frame called crimeOccurance that contains only one column from the data set called "Category"
crimeOccurance = df['Category']

## Prints the frequency of each category of crime.
print '\n\n'
print crimeOccurance.value_counts()

print '\n\n'

## Creates bar plot using the frequency count of each type of category. Plots Y label and title. 
crimeOccurance.value_counts().plot(kind='bar')

## Label the graph.
plt.ylabel('Frequency of Crime')
plt.xlabel('Category of Crime')
plt.title('Overview of Crime in the City of San Francisco\n')

## Reset the plot size.
fig = plt.gcf()
fig.set_size_inches(16.5, 10)
```

## Most Dangerous Neighbourhood in the City

Southern district of the city is the most dangerous area. The rate of crime in the southern district is almost 4 times as much as the safest district of the city. The crime rate in the southern district is almost one and half time as much as the next most unsafe district of the city Mission. Hence, it could be inferred that the police should do more patrolling in the southern district of the city which can help to bring down the crime rate drastically in the city.

The rate of the crime in each of the district is described in the table given below. The visualization after the table illustrates the tabular data in the bar chart format. Y axis of the graph represents the frequency of the crime whereas the X axis represents the district of the city.

```python
## Create a new Data frame called District that contains only one column from the data set called "PdDistrict"
district = df['PdDistrict']

## Creates bar plot using the frequency count of each type of category according to district. Plots Y label and title. 
print '\n\n'
print district.value_counts()
print '\n\n'

## Plot the Graph.
district.value_counts().plot(kind='bar',color='c')

## Label the graph.
plt.ylabel('Frequency of Crime')
plt.xlabel('Name of the District')
plt.title('Most Dangerous Neighbourhood in the City of San Francisco\n')

## Reset the size of the plot.
fig = plt.gcf()
fig.set_size_inches(15.5, 8.5)
```

## Criminal Activities across the San Francisco

Everyday across the San Francisco many different crimes is committed in various places. The visualization below illustrates the location and type of crime that has been occurred across the city. Map of the San Francisco can be seen below with proper coordinate points where X axis denotes latitude and y axis represents the longitude coordinate of the location. Each category has different color associated with the crime. It can be inferred from the visualization that theft, warrants and weapon law are most common type of offence across the city. If the police department of the city focuses on preventing such type of offence then the crime rate can be improved reasonably well. 

**Note:** Better the computer graphics performance better the colors would be represented. The are closely related due to the large number of categories of the crime.

```python
## Group by the category of the crime.
X = df[['Category','X','Y']]
groups = X.groupby(['Category'])

## For different colors across the scatter plots.
colors = cm.rainbow(np.linspace(0, 1, len(groups)))

## Plot each individual point and label it. Also give X and Y axis the label and title of the graph.
for group,c in zip(groups,colors):
    plt.plot(group[1]['X'], group[1]['Y'], marker='o', color=c, linestyle='', ms=12, label=group[0])

## Move the legend out of the plot.
plt.legend(bbox_to_anchor=(1.15, 1.05),prop={'size':10})

## Label the axis and put title of the graph.
plt.title('Location of Crime in City of San Francisco')
plt.xlabel('Latitude')
plt.ylabel('Longitude')

## Set the size of the plot.
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
```

## Week of Crime in the City of San Francisco

Generally speaking people associated with crime pick one day over other if they are looking to commit a big crime. The visualization below represents the number of offences that takes place on a particular day of the week. The days are sorted by the number of crimes that occurred on a given day. It can be inferred from the the bar graph that on fridays the crime activity of the city goes up. Although, there is not a huge difference between the number of crimes that take place on a particular day of the week; visualization manages to give good insight regarding the behaviour of criminal.

Data plotted in the visualization is described in the table given below Python code. Immediately after the code, bar chart representing the tabular data is plotted. 

```python
## Create a new Data frame called dayOfWeek that contains only one column from the data set called "DayOfWeek"
dayOfWeek = df['DayOfWeek']

## Creates bar plot using the frequency count of each type of category according to weekday. Plots Y label and title.
print '\n\n'
print dayOfWeek.value_counts()
print '\n\n'

## Plots the bar chart.
dayOfWeek.value_counts().plot(kind='bar',color='violet')

## Label the graph.
plt.ylabel('Frequency of Crime')
plt.xlabel('Day of the Week')
plt.title('Week of Crime in the City of San Francisco\n')

## Reset the size of the plot.
fig = plt.gcf()
fig.set_size_inches(15.5, 8.5)
```

## Day of Crime in the City of San Francisco

Some time of the days are very prone to crime while others are quite. Below is a bar graph that illustrates the number of crime that takes place in the city of San Francisco at any given hour. Highest number of offences take place between 5 PM to 7 PM whereas least amount of crime takes place between 4 AM to 5 AM.

Visualization below is also illustrated in tabular format above the bar chart. 

```python
## Format the dataframe
data = df.Time
data['Hour'] = data.str.split(':')

## Format the index column of the dataframe.
for i in range(0,data.size-1):
    hour = int(data['Hour'][i][0])
    if int(data['Hour'][i][0])<9:
        data['Hour'][i] = data['Hour'][i][0] +'-0'+str(hour+1)
    elif hour==23:
        data['Hour'][i] = data['Hour'][i][0] +'-0'+str(0)
    else:
        data['Hour'][i] = data['Hour'][i][0] +'-'+str(hour+1)
        
## Create new dataframe to process and create graph
graphD = data['Hour'].to_frame()
print '\n\n'
print graphD.groupby('Time').count()
print'\n\n'

## Plot bar chart using the data that were cleaned.
graphD.groupby('Time').count().plot(kind='bar',color='g')

## Label the graph.
plt.ylabel('Frequency of Crime')
plt.xlabel('Time of the Day')
plt.title('Day of Crime in San Francisco\n')

## Reset the size of the plot.
fig = plt.gcf()
fig.set_size_inches(17.5, 10)
```
